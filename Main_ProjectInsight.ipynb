```
from pytrends.request import TrendReq
```

```
pytrend = TrendReq(hl='en-IN', tz=360)
kw_list = [
  "Part time jobs",
  "Online Jobs",
  "Linkden",
  "Resume",
  "Naukri",

  "Economic crisis",
  "Crisis",
  "Recession",
  "Mandi",
  "Financial crisis",

  "NREGA",
  "Moratorium",
  "EMI",
  "ESIC",
  "Ration",

  "Bankruptcy",
  "Insolvency",
  "Loan Default",

  "Gold Loans",
  "Interest rate",
  "Home loan",
  "Car loan",
  "Personal loans",

  "Cement price",
  "Real Estate",
  "Majdoor",
  "Flats ",
  "Steel prices",

  "Mba",
  "Coursera",
  "Online classes",
  "Tuition",
  "Certification",

  "Ndtv",
  "Stock market",
  "Business news",
  "Elections",
  "cricket",

  "Irctc",
  "Goa",
  "App download",
  "Comedy ",
  "Lottery",

  "Jewellery",
  "Wedding",
  "Petrol price",
  "Cinema",
  "power cut",
]
```

```
len(kw_list)
```

```
import time
def fetch_trends(kw_list):
  dataset = []

  for kw in kw_list: 
    try:
      print("processing: " + kw)
      pytrend.build_payload(
          kw_list=[kw], 
          cat=0, 
          timeframe='2012-01-01 2020-12-01', 
          geo="IN")
      data = pytrend.interest_over_time()
      if not data.empty:
          data = data.drop(labels=['isPartial'], axis='columns')
          # data.set_index('date')
          dataset.append(data)
      time.sleep(5)
    except Exception as e:
      print("An exception occurred: " + kw)
      print(e)
  return dataset
```

```
batch1 = fetch_trends(kw_list)
```
