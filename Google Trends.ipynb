```python
import pandas as pd
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import SGD
from tensorflow.keras import regularizers
from tensorflow.keras import backend as K
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import numpy as np
import matplotlib.pyplot as plt
import shap
```

```python
from pytrends.request import TrendReq
```

```python
pytrend = TrendReq(hl='en-IN', tz=360)
kw_list = [
  "Part time jobs",
  "Online Jobs",
  "Linkden",
  "Resume",
  "Naukri",

  "Economic crisis",
  "Crisis",
  "Recession",
  "Mandi",
  "Financial crisis",

  "NREGA",
  "Moratorium",
  "EMI",
  "ESIC",
  "Ration",

  "Bankruptcy",
  "Insolvency",
  "Loan Default",

  "Gold Loans",
  "Interest rate",
  "Home loan",
  "Car loan",
  "Personal loans",

  "Cement price",
  "Real Estate",
  "Majdoor",
  "Flats ",
  "Steel prices",

  "Mba",
  "Coursera",
  "Online classes",
  "Tuition",
  "Certification",

  "Ndtv",
  "Stock market",
  "Business news",
  "Elections",
  "cricket",

  "Irctc",
  "Goa",
  "App download",
  "Comedy ",
  "Lottery",

  "Jewellery",
  "Wedding",
  "Petrol price",
  "Cinema",
  "power cut",
]
```

```python
len(kw_list)
```

```python
import time
def fetch_trends(kw_list):
  dataset = []

  for kw in kw_list: 
    try:
      print("processing: " + kw)
      pytrend.build_payload(
          kw_list=[kw], 
          cat=0, 
          timeframe='2012-01-01 2020-12-01', 
          geo="IN")
      data = pytrend.interest_over_time()
      if not data.empty:
          data = data.drop(labels=['isPartial'], axis='columns')
          # data.set_index('date')
          dataset.append(data)
      time.sleep(5)
    except Exception as e:
      print("An exception occurred: " + kw)
      print(e)
  return dataset
```

```python
batch1 = fetch_trends(kw_list)
```

```python
batch2 = fetch_trends(["Insolvency", "Loan Default"])
```

```python
result_df = pd.concat(batch1 + batch2, axis=1, join='inner')
result_df
```

```python
result_df.to_csv('final_trend.csv')
```

# Load Google tend

```python
#load DF from CSV
df = pd.read_csv('final_trend.csv')
excel_df = pd.read_excel('RBI Data.xlsx')
```

```python
df = df[3:]
excel_df = excel_df[:-1]
```

```python
df.reset_index(inplace = True,drop = True)
excel_df.reset_index(inplace = True, drop = True)
```

```python
df
```

```python
excel_df
```

```python
df['IIP'] = excel_df['IIP']
df['date'] = pd.to_datetime(df['date'])
```

```python
X = df.drop(['IIP','date'],axis=1)
y = df['IIP']
X_train,X_test,y_train,y_test = train_test_split(X,y,train_size=0.7,shuffle=False)
```

```python
normalizer = preprocessing.Normalizer().fit(X_train)
X_train_norm = normalizer.fit_transform(X_train)
X_test_norm = normalizer.fit_transform(X_test)
```

```python
model = Sequential()
model.add(Dense(100, input_shape=(50,),activation = 'relu'))
model.add(Dense(20,activation = 'relu'))
model.add(Dense(1,activation = 'relu'))
model.compile(optimizer = SGD(lr = 0.001), loss = 'mean_squared_error', metrics = ['mean_squared_error'])

```

```python
network = model.fit(X_train_norm, y_train, epochs = 9000, batch_size = 26, verbose = 1)
```

```python
prediction = model.predict(X_test_norm, verbose = 1)
```

```python
prediction
```

```python
print(mean_squared_error(y_test,prediction))
```

```python
e = shap.DeepExplainer(model,X_train)
shap_values = e.shap_values(model,X_train)
shap.summary_plot(shap_values, X_train)
```
